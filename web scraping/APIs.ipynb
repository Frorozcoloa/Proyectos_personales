```python
def get_token(client_id, client_secret):
    """
    Funci칩n que extrae un JSON con la informaci칩n de un album identificado con id
    """
    encoded = base64.b64encode(bytes(client_id+':'+client_secret, 'utf-8'))
    params = {'grant_type' : 'client_credentials'}
    header = {'Authorization' : 'Basic ' + str(encoded, 'utf-8')}
    r = requests.post('https://accounts.spotify.com/api/token', headers=header, data=params)
    if r.status_code != 200:
        print('Error en la request', r.json())
        
        return None
    else:
        print(f'Token v치lido por {r.json()["expires_in"]} segundos')
        return r.json()['access_token']
```

```python
def obtener_discografia(artista_id, token, return_name = False, page_limit = 50, country = None):
    url = 'https://api.spotify.com/v1/artists/{}/albums'
    header = {'Authorization': f'Bearer {token}'}
    params = {'limit' : page_limit,
             'offset' : 0,
             'country': country}
    
    lista = []
    r = requests.get(url.format(artista_id), params = params, headers = header)
    if r.status_code != 200:
        print('Error en la request tipo 1', r.json())
        return None
    
    if return_name:
        lista += [(item['id'], item['name']) for item in r.json()['items']]
        
    else:
        lista += [(item['id']) for item in r.json()['items']]
    

    while r.json()['next']:
        # print(f'oleada: {i}')
        r = requests.get(r.json()['next'], headers=header)
        
        if r.status_code != 200:
            print('Error en la request tipo 2', r.json())
            return lista
        if return_name:
            lista += [(item['id'], item['name']) for item in r.json()['items']]
        else:
            lista += [(item['id']) for item in r.json()['items']]
       
    return lista
    
    
```

```python
def obtener_tracks(album_id, token, return_name = False, page_limit = 50, market = None):
    url = 'https://api.spotify.com/v1/albums/{}/tracks'
    header = {'Authorization': f'Bearer {token}'}
    
    params = {'limit' : page_limit,
             'offset' : 0,
              'market' : market
              }
    
    
    lista = []
    
    r = requests.get(url.format(album_id), params = params, headers = header)
    if r.status_code != 200:
        print('Error en la request tipo 1', r.json())
        return None
    
    if return_name:
        lista += [(item['id'], item['name']) for item in r.json()['items']]
        
    else:
        lista += [(item['id']) for item in r.json()['items']]
    

    
    while r.json()['next']:
        r = requests(r.json()['next'], headers=header)
        
        if r.status_code != 200:
            print('Error en la request tipo 2', r.json())
            return lista
        
        if return_name:
            lista += [(item['id'], item['name']) for item in r.json()['items']]
        
        else:
            lista += [(item['id']) for item in r.json()['items']]
    return lista
```
